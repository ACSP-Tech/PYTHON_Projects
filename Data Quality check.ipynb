```python
#importing the necessary requirement
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#importing the data into pandas dataframe
url = 'https://github.com/ACSP-Tech/DATA-/raw/main/KPMG_VI_New_raw_data_update_final%20(4).xlsx'
sprocket_df = pd.ExcelFile(url, engine='openpyxl')
print('Excel sheet = ', sprocket_df.sheet_names)

#parse all the sheet to excel dataframe
Transactions = sprocket_df.parse(sheet_name= 'Transactions', skiprows=1)
CustomerDemographic = sprocket_df.parse(sheet_name= 'CustomerDemographic', skiprows=1)
NewCustomerList = sprocket_df.parse(sheet_name= 'NewCustomerList', skiprows=1)
CustomerAddress = sprocket_df.parse(sheet_name= 'CustomerAddress', skiprows=1)

print('\n\n', 'Transactions Data exploration:', '\n\n')
print(Transactions.info())
print(Transactions.describe())
for col in Transactions.columns:
    unique_values = Transactions[col].unique()
    print(f"Column '{col}': Unique values = {unique_values}")
Transactions.head()
```
